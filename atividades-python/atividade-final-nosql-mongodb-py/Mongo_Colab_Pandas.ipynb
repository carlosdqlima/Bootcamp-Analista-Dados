```python
from bson.objectid import ObjectId
from pymongo import MongoClient
import pandas as pd

# Conectar ao MongoDB Atlas usando a string de conexão com a senha correta
cliente = MongoClient("mongodb+srv://carlos:qPNWv5RPbEFmdcZP@meu-mongodb.imbfr.mongodb.net/?retryWrites=true&w=majority&appName=meu-mongodb")

def menu():
  print("------------------------MENU------------------------")
  print("1 - Inserir livro(s)")
  print("2 - Listar todos os livros cadastrados")
  print("3 - Buscar livros por título, autor ou ISBN")
  print("4 - Atualizar livro(s)")
  print("5 - Deletar livro(s)")
  print("6 - Sair")
  print("----------------------------------------------------")

def create():
  inserir_livros = {
      'titulo': input("Digite o titulo do livro: "),
      'autor': input("Digite o nome do autor desse livro: "),
      'ano': int(input("Digite o ano de lançamento do livro: ")),
      'isbn': input("Digite o ISBN do livro: ")}
  db.livros.insert_one(inserir_livros)
  print("----------------------------------------------------")
  print("Livro inserido com sucesso!")
  print("----------------------------------------------------")

def read():
  livros = collection.find()
  data = list(livros)
  df = pd.DataFrame(data)
  display(df)
  print("----------------------------------------------------")
  print("Livros listados com sucesso!")
  print("----------------------------------------------------")

def update():
  livro_id = input("Digite o ID do livro que deseja atualizar: ")
  atualizar_livros={
      'titulo': input("Digite o novo titulo do livro: "),
      'autor': input("Digite o novo nome do autor desse livro: "),
      'ano': int(input("Digite o novo ano de lançamento do livro: ")),
      'isbn': input("Digite o novo isbn do livro: ")
      }

  db.livros.update_one(
      {"_id": ObjectId(livro_id)},
      {"$set": atualizar_livros}
  )

  print("----------------------------------------------------")
  print("Livro atualizado com sucesso!")
  print("----------------------------------------------------")

def delete():
  livro_id = input("Digite o ID do livro que deseja deletar: ")
  db.livros.delete_one({"_id": ObjectId(livro_id)})
  print("----------------------------------------------------")
  print("Livro deletado com sucesso!")
  print("----------------------------------------------------")

def buscar_livro():
  print("Digite o critério de busca:")
  print("1 - Título")
  print("2 - Autor")
  print("3 - ISBN")
  print("4 - Voltar ao menu principal")
  print("----------------------------------------------------")
  opcao = int(input("Digite a opção desejada: "))
  print("----------------------------------------------------")
  if opcao == 1:
    titulo = input("Digite o titulo do livro a ser pesquisado: ")
    livros = collection.find({"titulo": (titulo)})
    data = list(livros)
    df = pd.DataFrame(data)
    print("----------------------------------------------------")
    print("Livro encontrado com sucesso!")
    print("----------------------------------------------------")
    display(df)
  elif opcao == 2:
    autor = input("Digite o autor do livro a ser pesquisado: ")
    livros = collection.find({"autor": (autor)})
    data = list(livros)
    df = pd.DataFrame(data)
    print("----------------------------------------------------")
    print("Livro encontrado com sucesso!")
    print("----------------------------------------------------")
    print(df)
  elif opcao == 3:
    isbn = input("Digite o ISBN do livro a ser pesquisado: ")
    livros = collection.find({"isbn": (isbn)})
    data = list(livros)
    df = pd.DataFrame(data)
    print("----------------------------------------------------")
    print("Livro encontrado com sucesso!")
    print("----------------------------------------------------")
    display(df)
  elif opcao == 4:
    menu()
  else:
    print("----------------------------------------------------")
    print("Opção inválida. Tente novamente.")
    print("----------------------------------------------------")

def main():
  while True:
    menu()
    opcao = int(input("Digite a opção desejada: "))
    print("----------------------------------------------------")
    if opcao == 1:
      create()
    elif opcao == 2:
      read()
    elif opcao == 3:
      buscar_livro()
    elif opcao == 4:
      update()
    elif opcao == 5:
      delete()
    elif opcao == 6:
      print("-------------------------------------------------------")
      print("Obrigado por usar o Sistema de Gerenciamento de Livros!")
      print("-------------------------------------------------------")
      break
    else:
      print("----------------------------------------------------")
      print("Opção inválida. Tente novamente.")
      print("----------------------------------------------------")

print("-------------------------------------------------------")
print("Seja bem-vindo ao Sistema de Gerenciamento de Livros")

main()
```
